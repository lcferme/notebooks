```python
import pyclesperanto_prototype as cle


image = cle.create((100,100,100))
cle.set(image, 0)

image = cle.draw_sphere(image, x=50, y=50, z=50, radius_x=20, radius_y=10, radius_z=50, value=1)
cle.imshow(image)

cle.imshow(cle.transpose_xz(image))
cle.imshow(cle.transpose_yz(image))
```

```python
from skimage.measure import regionprops, regionprops_table

stats = regionprops(cle.pull(image).astype(int))
```

```python
print(stats[0]['minor_axis_length'])
print(stats[0]['major_axis_length'])
```
